```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#loading the dataset
cancer_dataset= sklearn.datasets.load_breast_cancer()
```

```python
print(cancer_dataset)
```

```python
#load data into a dataframe
data_frame=pd.DataFrame(cancer_dataset.data, columns=cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
#adding target column to the dataframe
data_frame['label']=cancer_dataset.target
data_frame.tail()
```

```python
data_frame.info()
```

```python
data_frame.describe()
```

```python
data_frame.shape
```

```python
data_frame['label'].value_counts()
```

```python
#splitting the data
x=data_frame.drop(columns='label',axis=1)
y=data_frame['label']
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=2)
print(x.shape,x_train.shape,x_test.shape)
print(y.shape,y_train.shape,y_test.shape)
```

STANDARIZING THE DATA

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar= StandardScaler()
x_train_std=scalar.fit_transform(x_train)
x_test_std=scalar.transform(x_test)
```

```python
print(x_train_std)
```

BUILDING A NEURAL NETWORK

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

no of neurons in input layer =no of features in the data

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),# (input layer) only no of features, dont include target column
    keras.layers.Dense(20,activation='relu'),# (hidden layer)rectified linear units
    keras.layers.Dense(2, activation='sigmoid'),# (output layer) no of neurons=no of classes, in this case 1 or 0
])
```

```python
#compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

dog-0, cat-1, horse-2
This is known as integer representation. In this case we use SparseCategiricalCrossentropy

dog-[1 0 0], cat-[0 1 0] , horse-[0 0 1]
this is known as one hot representation. In this case we use CategoricalCrossentropy


```python
#training the neural network
history=model.fit(x_train_std,y_train,validation_split=0.1,epochs=10)# epoch refers to no of times training dataset passes through the algorithm
```

VISUALIZATION OD ACCURACY AND LOSS


```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("model's accuracy")
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training_data','validation_data'], loc='upper right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title("model's loss")
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training_data','validation_data'], loc='lower right')
plt.show()
```

```python
#on test data
loss, accuracy= model.evaluate(x_test_std,y_test)
print(accuracy)
print(loss)
```
